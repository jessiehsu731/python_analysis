```python
#Matplotlib https://matplotlib.org/stable/api/index.html
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')
```

```python
figure = plt.figure(figsize=(10, 6 ))
axes = figure.add_subplot(1,1,1)
axes.set_xticks([1,5,6,2,9])
axes.set_yticks([-1,2,0,0.5,2])
plt.show()
```

```python
figure = plt.figure(figsize=(10, 6 ))
axes = figure.add_subplot(1,1,1)
axes.plot([11,3,6,7,9],marker='o', color='red', linestyle='-.')
plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
ax1 = figure.add_subplot(2,1,1)
ax1.plot([11, 3, 5, 7, 9],marker='o',color='red', linestyle='')
ax2 = figure.add_subplot(2,1,2)
ax2.bar(['a', 'b', 'c', 'x', 'y'],[1, 2, 3, 4, 5])
plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
ax1 = figure.add_subplot(3,1,1)
ax1.plot([11, 3, 5, 7, 9], marker='o', color='green', linestyle='--')
ax2 = figure.add_subplot(3,1,2)
ax2.bar(['a', 'b', 'c', 'x', 'y'],[1, 2, 3, 4, 5], color='lightblue')
ax3 = figure.add_subplot(3,1,3)
ax3.plot(['1', '2', '3', '4', '5'],[0, 0.5, 0.6, 0.4, 0.7], marker='x', color='red', linestyle='--')
plt.show()
```

```python
student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects, score1, marker='x', color='red', linestyle='-.')
axes.plot(subjects, score2, marker='o', color='green', linestyle='--')
plt.show()

```
