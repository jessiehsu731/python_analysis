```python
student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.set_title('小名和大名分數比較表')
axes.set_xlabel('科目')
axes.set_ylabel('分數')
axes.plot(subjects, score1, 'xr-',label="小名")
axes.plot(subjects, score2, 'og-.',label="大名")
axes.legend()
axes.grid(axis='y')

#加入標註，xy是要標註的位置，xytext是標註文字的位置, arrowprops是箭頭的設定, textcoords是文字的座標
axes.annotate(f"{student1['國文']}", xy=(subjects[0], score1[0]), xytext=(subjects[0], score1[0]+5), 
              arrowprops=dict(facecolor='gray', shrink=0.05))
plt.show()
```

```python
subjects, score1, score2  #tuple顯示
print(subjects)
print(score1)
print(score2)
```
