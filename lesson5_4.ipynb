###讀取csv

```python
import pandas as pd
df = pd.read_csv('student.csv',index_col=0)  # 利用index_col=0 來指定第一個欄位作為索引
df
```

###讀取excel

```python
df = pd.read_excel('student.xlsx',index_col=0) 
df
df.head()
```

###透過欄選取
###透過列選取

```python
#透過欄選取
#單欄選取
df['chinese']  #一維資料--series
```

```python
#多欄選取
df[['chinese','math']]    #subscript--最外圍[]叫subscript
```

```python
#loc->使用索引的名稱
#iloc->使用索引的index

df.loc['王坤安']
df.loc[['王坤安','林俊翔']]  #裡面再放一個[]代表要選多個
```

```python
df.loc['林俊翔':'王坤安'] 
df.iloc[-5:-1] 
df.iloc[0:2] 
```

```python
df.head()  #可以看前面五筆
df.loc['柯韋成']   #series只有索引沒有欄位
#df.loc['柯韋成']['geo'] = 70
#df

df.loc['柯韋成','geo'] = 70  #如果用loc索引，index後面column，中間可以直接放逗號就好
df
df.head()
```
