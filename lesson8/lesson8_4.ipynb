```python
import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv')
df.info
```

```python
df1 = df.dropna()
df1.info()
df1.tail()
df1.head()
```

```python
df2 = df1.rename(columns = df.iloc[0,:])
df2

df2 = df1.rename(columns=df1.iloc[0].to_dict())
df2
```

```python
df3 = df2.drop(index=0)
df3.info()
```

```python
import numpy as np
def str_to_int(value:str) -> int:
    try:
        value_int = int(value)
    except:
        value_int = 0
    return value_int

def str_to_float(value:str) -> float:
    try:
        value_float = float(value)
    except:
        value_float = 0
    return value_float

df3['年底人口數'] = df3['年底人口數'].map(str_to_int)
df3['人口密度'] = df3['人口密度'].map(str_to_float)
df3['土地面積'] = df3['土地面積'].map(str_to_float)

df3.info()
```

```python
df3
```

```python
def split(value:str) -> str:
    if len(value) > 4:
        return value[:3]
    else:
        return value
    
df3['縣市'] = df3['區域別'].map(split)
df3
```

```python
df4 = df3.reindex(columns=["縣市","區域別","年底人口數","土地面積","人口密度"])
df4
```

```python
def population(value:int) -> str:
    if value > 367653:
        return '人多'
    elif value > 183826:
        return '適中'
    else:
        return '人多'
df4['人口分布'] = df4['年底人口數'].map(population)
df4
```

```python
pd.cut(df4['年底人口數'],bins=3)#分三類categogy
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   #seaborn在彌補matplotlib不足的地方

df_plot = df4[['縣市','年底人口數']]

plt.figure(figsize=(12,6))
sns.boxplot(data=df_plot, x='縣市', y='年底人口數')
plt.xlabel('區域別')
plt.ylabel('年底人口數')
plt.title('各區域人口數量盒鬚圖')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df_plot, x='年底人口數', hue='縣市', multiple="stack", kde=False)
plt.xlabel('年底人口數')
plt.ylabel('次數')
plt.title('各縣市人口數量分布圖')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
