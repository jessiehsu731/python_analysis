```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 101, size=(50,5))
scores

df = pd.DataFrame(scores,
             columns = ['國文','英文','數學','地理','歷史'],
             index=range(1,51)
             )
df.index.name = '學號'
df.columns.name = '科目'
df
```

```python
sum_value = df.sum(axis=1)
mean_value = df.mean(axis=1)
```

```python
df['總分'] = sum_value
df['平均'] = mean_value
df
```

```python
#排序
df1 = df.sort_values(by='總分', ascending=False) #高到低
df1
```

```python
rank_value = df1['總分'].rank(ascending=False, method='min')  #ascending從排名1-50   #用min這樣同排名的會自動變同排名，不會是小數點
df1['排名'] = rank_value
df1
```

```python
def max_score(row) -> int:
    return row.max()   #傳出最高分

#抓每個人最高科目分數
df1[['國文','英文','數學','地理','歷史']].apply(max_score, axis=1)
```

```python
def min_score(row) -> int:
    return row.min()

df1[['國文','英文','數學','地理','歷史']].apply(min_score, axis=1)
```

```python
def score_list(row) -> int:
    less_60_count = len(row[row < 60])
    return pd.Series([row.min(),row.max(),row.std(),less_60_count],index=['最低分','最高分','標準差','不及格的數量'])

df1[['國文','英文','數學','地理','歷史']].apply(score_list, axis=1)
```
